```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Opponent": ["RR", "DC", "CSK", "PBKS", "LSG"],
    "SRH_Score": [229, 242, 194, 219, 156],
    "Opponent_Score": [228, 195, 184, 223, 160],
    "Result": ["Win", "Win", "Win", "Loss", "Loss"],
    "Best_Player": [
        "Abhishek Sharma",
        "Travis Head",
        "Heinrich Klaasen",
        "Shikhar Dhawan",
        "KL Rahul"
    ]
}

df = pd.DataFrame(data)
df
```

```python
print("Average Score:", df["SRH_Score"].mean())
print("Highest Score:", df["SRH_Score"].max())
print("Total Wins:", (df["Result"] == "Win").sum())
```

```python
#bar chart
plt.figure()
plt.bar(df["Opponent"], df["SRH_Score"],color='green')
plt.title("SRH Scores vs Opponents")
plt.xlabel("Opponent")
plt.ylabel("Runs")
plt.show()
```

```python
#pie chart
result_counts = df["Result"].value_counts()

plt.figure()
plt.pie(result_counts, labels=result_counts.index, autopct='%1.1f%%',explode=(0.2,0.1))
plt.title("Match Results Distribution")
plt.show()
```

```python
#line chart
plt.figure()
plt.plot(df["Opponent"], df["SRH_Score"], marker='o', label="SRH")
plt.plot(df["Opponent"], df["Opponent_Score"], marker='o', label="Opponent")
plt.title("Score Comparison")
plt.xlabel("Opponent")
plt.ylabel("Runs")
plt.legend()
plt.show()
```

```python
#runs and score
plt.figure()
plt.bar(["Highest", "Lowest"],
        [df["SRH_Score"].max(), df["SRH_Score"].min()],color='pink')
plt.title("Highest vs Lowest Score")
plt.ylabel("Runs")
plt.show()
```

```python
#win vs loss
win_loss = df["Result"].value_counts()

plt.figure()
plt.bar(win_loss.index, win_loss.values,color='yellow')
plt.title("Wins vs Losses")
plt.ylabel("Matches")
plt.show()
```

```python
#win%
win_percentage = (df["Result"].value_counts()["Win"] / len(df)) * 100
print("Win Percentage:", win_percentage)
```

```python
#best player
best_player_count = df["Best_Player"].value_counts()
print(best_player_count)
```

```python
#Best Player Frequency
plt.figure()
plt.bar(best_player_count.index, best_player_count.values,color='cyan')
plt.title("Best Player Count")
plt.xlabel("Player")
plt.ylabel("Number of Matches")
plt.xticks(rotation=30)
plt.show()
```
